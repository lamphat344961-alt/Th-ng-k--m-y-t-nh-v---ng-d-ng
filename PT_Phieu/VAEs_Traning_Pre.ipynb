* **Ho·∫°t ƒë·ªông h∆∞·ªõng nghi·ªáp:** C√°c ho·∫°t ƒë·ªông di·ªÖn ra trong th·ªùi gian ƒëi h·ªçc.
* **Y·∫øu t·ªë ·∫£nh h∆∞·ªüng ngh·ªÅ nghi·ªáp:** ƒêo l∆∞·ªùng b·∫±ng thang ƒëo Likert.

    * G·ªìm 15 c√¢u h·ªèi th√†nh ph·∫ßn: `q12_rate1` ƒë·∫øn `q12_rate15`.
    * **Bi·∫øn t·ªïng h·ª£p:** `tong12` (Gi√° tr·ªã trung b√¨nh c·ªßa 15 c√¢u h·ªèi tr√™n).

---

## 3. üéØ Bi·∫øn m·ª•c ti√™u (Target Variable)
* **T√™n bi·∫øn:** `p1q9_job_current`
* **M√¥ t·∫£:** Bi·∫øn ph√¢n lo·∫°i g·ªìm 6 nh√≥m ngh·ªÅ (m√£ h√≥a t·ª´ `0` ƒë·∫øn `5`).
* **‚ö†Ô∏è L∆∞u √Ω v·ªÅ ph√¢n ph·ªëi l·ªõp:** D·ªØ li·ªáu b·ªã **m·∫•t c√¢n b·∫±ng nghi√™m tr·ªçng (Imbalanced Data)**.
    * **Nh√≥m 0:** Chi·∫øm ∆∞u th·∫ø (> 50% m·∫´u).
    * **Nh√≥m 1‚Äì3:** S·ªë l∆∞·ª£ng trung b√¨nh.
    * **Nh√≥m 4‚Äì5:** R·∫•t √≠t quan s√°t (ƒë·∫∑c bi·ªát Nh√≥m 5 chi·∫øm < 1%).

---

## 4. ƒê·∫∑c ƒëi·ªÉm x·ª≠ l√Ω d·ªØ li·ªáu

### ‚ö†Ô∏è M√£ h√≥a ƒë·∫∑c bi·ªát (Special Encoding)
Trong t·∫≠p d·ªØ li·ªáu xu·∫•t hi·ªán c√°c gi√° tr·ªã √¢m mang √Ω nghƒ©a ƒë·ªãnh danh, **kh√¥ng c√≥ √Ω nghƒ©a s·ªë h·ªçc**:
* `-1`: Kh√¥ng √°p d·ª•ng.
* `-2`: Kh√¥ng tr·∫£ l·ªùi.
> **L∆∞u √Ω:** Ph·∫£i x·ª≠ l√Ω c√°c gi√° tr·ªã n√†y tr∆∞·ªõc khi ƒë∆∞a v√†o m√¥ h√¨nh t√≠nh to√°n.

### ‚ÑπÔ∏è Thang ƒëo h√†i l√≤ng (Satisfaction Scale)
* 15 bi·∫øn (`q12_rate1` ‚Äì `q12_rate15`) t·∫°o th√†nh m·ªôt thang ƒëo nh·∫•t qu√°n (Cronbach's alpha cao).
* Bi·∫øn `tong12`: Ph·∫£n √°nh k·∫øt qu·∫£ **sau khi ƒë√£ ƒëi l√†m**.
> **Quan tr·ªçng:** Bi·∫øn `tong12` l√† k·∫øt qu·∫£ (outcome), **kh√¥ng ƒë∆∞·ª£c s·ª≠ d·ª•ng l√†m bi·∫øn nguy√™n nh√¢n (feature)** ƒë·ªÉ d·ª± b√°o ngh·ªÅ nghi·ªáp hi·ªán t·∫°i ƒë·ªÉ tr√°nh r√≤ r·ªâ d·ªØ li·ªáu (data leakage).

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
import pandas as pd
import pickle
from sklearn.model_selection import StratifiedKFold
from collections import defaultdict
from xgboost import XGBClassifier
```

# üßπ Ti·ªÅn X·ª≠ L√Ω: Lo·∫°i B·ªè Bi·∫øn (Feature Selection)

ƒê·ªÉ ƒë·∫£m b·∫£o t√≠nh ch√≠nh x√°c c·ªßa m√¥ h√¨nh v√† tr√°nh hi·ªán t∆∞·ª£ng **R√≤ r·ªâ d·ªØ li·ªáu (Data Leakage)**, c√°c bi·∫øn sau ƒë√¢y s·∫Ω ƒë∆∞·ª£c lo·∫°i b·ªè kh·ªèi t·∫≠p d·ªØ li·ªáu hu·∫•n luy·ªán.

Nh·ªØng bi·∫øn n√†y ph·∫£n √°nh tr·∫°ng th√°i **sau khi ƒë√£ ƒëi l√†m** ho·∫∑c l√† **c·∫£m nh·∫≠n hi·ªán t·∫°i**, do ƒë√≥ kh√¥ng th·ªÉ d√πng l√†m ƒë·∫ßu v√†o ƒë·ªÉ d·ª± b√°o (v·ªën dƒ© ph·∫£i l√† d·ªØ li·ªáu c√≥ tr∆∞·ªõc khi s·ª± ki·ªán x·∫£y ra).

| Nh√≥m th√¥ng tin | C√¢u h·ªèi / M√¥ t·∫£ bi·∫øn | L√Ω do lo·∫°i b·ªè |
| :--- | :--- | :--- |
| **ƒê√°nh gi√° hi·ªán t·∫°i** | *"T√≠ch v√†o nh·ªØng c√¢u tr·∫£ l·ªùi sau ƒë√¢y theo t·ª´ng m·ª©c ƒë·ªô m√† Anh/ch·ªã cho l√† ƒë√∫ng v·ªõi m√¨nh"* | Ph·∫£n √°nh t√¢m l√Ω hi·ªán t·∫°i, kh√¥ng ph·∫£i ƒë·∫∑c ƒëi·ªÉm c·ªë ƒë·ªãnh. |
| **C·∫£m x√∫c hi·ªán t·∫°i** | *"Hi·ªán t·∫°i, b·∫°n c·∫£m th·∫•y th·∫ø n√†o v·ªÅ l·ª±a ch·ªçn c·ªßa m√¨nh?"* | D·ªØ li·ªáu thu th·∫≠p sau khi s·ª± vi·ªác ƒë√£ x·∫£y ra (Post-hoc evaluation). |
| **L√Ω do ch·ªçn tr∆∞·ªùng** | *"H√£y cho bi·∫øt v√¨ sao Anh/ch·ªã quy·∫øt ƒë·ªãnh ch·ªçn tr∆∞·ªùng m√† anh/ch·ªã ƒë√£ thi..."* | D·ªØ li·ªáu d·∫°ng vƒÉn b·∫£n m·ªü (Text), kh√≥ ƒë·ªãnh l∆∞·ª£ng tr·ª±c ti·∫øp v√† mang t√≠nh h·ªìi t∆∞·ªüng. |
| **Li√™n h·ªá c√¥ng vi·ªác** | *"C√¥ng vi·ªác hi·ªán nay c·ªßa anh/ch·ªã c√≥ li√™n quan g√¨ ƒë·∫øn ngh·ªÅ ƒë∆∞·ª£c h·ªçc ·ªü THCS kh√¥ng?..."* | **Data Leakage:** Ti·∫øt l·ªô tr·ª±c ti·∫øp th√¥ng tin v·ªÅ c√¥ng vi·ªác hi·ªán t·∫°i. |
| **ƒê√°nh gi√° ch·ªß quan** | *"Ngh·ªÅ nghi·ªáp m√† anh/ch·ªã cho l√† th√†nh c√¥ng nh·∫•t nh·ªù vi·ªác h∆∞·ªõng nghi·ªáp ·ªü THCS?"* | Mang t√≠nh ch·ªß quan c√° nh√¢n v·ªÅ qu√° kh·ª©. |
| **L·ªãch s·ª≠ thay ƒë·ªïi** | *"Anh/ch·ªã ƒë√£ t·ª´ng thay ƒë·ªïi c√¥ng vi·ªác?"* | Th√¥ng tin x·∫£y ra trong qu√° tr√¨nh ƒëi l√†m (Future information). |
| **K·∫øt qu·∫£ ƒë·∫ßu ra** | *"Ngh·ªÅ nghi·ªáp c·ªßa anh/ch·ªã sau khi h·ªçc xong..."* | ƒê√¢y l√† k·∫øt qu·∫£ (Outcome), t∆∞∆°ng ƒë∆∞∆°ng v·ªõi bi·∫øn m·ª•c ti√™u. |
| **Bi·∫øn t·ªïng h·ª£p** | **Bi·∫øn `tong12`**: *"Trung b√¨nh nh·ªØng c√¢u tr·∫£ l·ªùi sau ƒë√¢y..."* | Bi·∫øn t·ªïng h·ª£p t·ª´ thang ƒëo h√†i l√≤ng sau khi ƒëi l√†m (Outcome variable). |

---
> **K·∫øt lu·∫≠n:** Ch·ªâ gi·ªØ l·∫°i c√°c bi·∫øn thu·ªôc v·ªÅ **Th√¥ng tin n·ªÅn (Demographics)** v√† **Y·∫øu t·ªë ·∫£nh h∆∞·ªüng/H∆∞·ªõng nghi·ªáp trong qu√° tr√¨nh ƒëi h·ªçc** ƒë·ªÉ x√¢y d·ª±ng m√¥ h√¨nh d·ª± b√°o.

```python

# ======================
# CONFIG
# ======================
DATA_PATH = "Dataset_Phieu_final.csv"
TARGET = "p1q9_job_current"


# ======================
# Helper
# ======================
def select_by_prefix(df, prefixes):
    return [c for c in df.columns if any(c.startswith(p) for p in prefixes)]


def select_existing(df, names):
    return [c for c in names if c in df.columns]


# ======================
# MAIN
# ======================
df = pd.read_csv(DATA_PATH)

if TARGET not in df.columns:
    raise ValueError(f"Kh√¥ng t√¨m th·∫•y c·ªôt target '{TARGET}'")

# ---------- KEEP: THCS-only ----------
keep_exact = [
    # Nh√¢n kh·∫©u h·ªçc / n·ªÅn t·∫£ng
    "p1q5_gender",
    "p1q6_ethn",
    "p1q2_typeschool",
    "p1q3_typeclass",

    # Gia ƒë√¨nh
    "p1q12_fjob",
    "p1q13_mjob",
    "p1q14_megr9",

    # M√¥n h·ªçc
    "p1q15_hsub",
    "p1q16_lsub",
    "p1q16_likesub",
    "p1q16_dislikesub",

    # H∆∞·ªõng nghi·ªáp
    "q1_trainjob",
    "q1a_namtrainjob",
    "q1c_cer",
    "q8_ins",
    "q9_insexam",
    "p1q8_afloseschool",
]

keep_prefixes = [
    "q2_care",
    "q3_where",
    "q4_act",
    "q5_affect",
    "q6_like",
    "requi",
    "aware",
]

keep_cols = []
keep_cols += select_existing(df, keep_exact)
keep_cols += select_by_prefix(df, keep_prefixes)

# b·ªè tr√πng
keep_cols = list(dict.fromkeys(keep_cols))


# ---------- DROP: leakage / kh√¥ng THCS-only ----------
drop_exact = [
    "p1q8a_job_afloseschool",
    "p1q10_chanjob",
    "tong12",
]

drop_prefixes = [
    "q12_rate",
    "q10_whycho",
]

keep_cols = [
    c for c in keep_cols
    if (c not in drop_exact)
    and (not any(c.startswith(p) for p in drop_prefixes))
]

# ---------- Build filtered dataset ----------
df_filtered = df[keep_cols + [TARGET]].copy()

# ---------- Save ----------
df_filtered.to_csv("dataset_THCS_only_filtered.csv", index=False)

# ---------- Quick check ----------
print("Rows:", df_filtered.shape[0])
print("Features kept:", len(keep_cols))
print("Target distribution:")
print(df_filtered[TARGET].value_counts())
```

# üîÑ Chi·∫øn L∆∞·ª£c Ki·ªÉm ƒê·ªãnh: Stratified K-Fold Cross-Validation

ƒê·ªÉ ƒë√°nh gi√° hi·ªáu nƒÉng m√¥ h√¨nh m·ªôt c√°ch kh√°ch quan nh·∫•t, ch√∫ng t√¥i √°p d·ª•ng ph∆∞∆°ng ph√°p **Ki·ªÉm ƒë·ªãnh ch√©o ph√¢n t·∫ßng (Stratified K-Fold Cross-Validation)**.

### 1. M·ª•c ti√™u
* **TƒÉng ƒë·ªô tin c·∫≠y (Robustness):** ƒê·∫£m b·∫£o k·∫øt qu·∫£ ƒë√°nh gi√° ph·∫£n √°nh ƒë√∫ng nƒÉng l·ª±c c·ªßa m√¥ h√¨nh, lo·∫°i b·ªè y·∫øu t·ªë "may m·∫Øn" do vi·ªác chia d·ªØ li·ªáu ng·∫´u nhi√™n g√¢y ra.
* **Gi·∫£m thi·ªÉu Overfitting:** Ki·ªÉm so√°t t·ªët h∆°n vi·ªác m√¥ h√¨nh h·ªçc v·∫πt tr√™n m·ªôt t·∫≠p train c·ª• th·ªÉ.

### 2. L√Ω do l·ª±a ch·ªçn
* D·ªØ li·ªáu hi·ªán t·∫°i ƒëang b·ªã **m·∫•t c√¢n b·∫±ng nghi√™m tr·ªçng (Imbalanced Class)**.
* Ph∆∞∆°ng ph√°p **Stratified** gi√∫p ƒë·∫£m b·∫£o t·ª∑ l·ªá gi·ªØa c√°c l·ªõp (classes) trong m·ªói fold (t·∫≠p con) t∆∞∆°ng ƒë·ªìng v·ªõi t·ª∑ l·ªá c·ªßa t·∫≠p d·ªØ li·ªáu g·ªëc, tr√°nh tr∆∞·ªùng h·ª£p m·ªôt fold ch·ªâ to√†n ch·ª©a nh√£n ƒëa s·ªë.

```python

# ======================
# CONFIG
# ======================
DATA_PATH = "dataset_THCS_only_filtered.csv"
OUT_DIR = Path("cv_outputs")
OUT_DIR.mkdir(parents=True, exist_ok=True)

TARGET = "p1q9_job_current"
N_SPLITS = 3
SEED = 42


# ======================
# LOAD DATA
# ======================
df = pd.read_csv(DATA_PATH)

X = df.drop(columns=[TARGET])
y = df[TARGET]

# ======================
# FULL CLASS DISTRIBUTION
# ======================
class_counts_full = (
    y.value_counts()
     .rename_axis("class")
     .reset_index(name="count")
     .sort_values("class")
)
# L∆∞u l·∫°i s·ªë l∆∞·ª£ng t·ª´ng l·ªõp
class_counts_full.to_csv(
    OUT_DIR / "class_counts_full.csv",
    index=False
)


# ======================
# STRATIFIED K-FOLD (3 folds)
# ======================
skf = StratifiedKFold(
    n_splits=N_SPLITS,
    shuffle=True,
    random_state=SEED
)

fold_indices = []
per_fold_counts = []

for fold_id, (train_idx, val_idx) in enumerate(skf.split(X, y)):
    fold_indices.append({
        "fold": fold_id,
        "train_idx": train_idx,
        "val_idx": val_idx,
    })

    # class distribution in validation fold
    fold_counts = (
        y.iloc[val_idx]
        .value_counts()
        .rename_axis("class")
        .reset_index(name="count")
    )
    fold_counts["fold"] = fold_id
    per_fold_counts.append(fold_counts)

# ======================
# SAVE FOLDS
# ======================
with open(OUT_DIR / "cv_folds.pkl", "wb") as f:
    pickle.dump(fold_indices, f)

class_counts_per_fold = (
    pd.concat(per_fold_counts, ignore_index=True)
      .sort_values(["fold", "class"])
)

class_counts_per_fold.to_csv(
    OUT_DIR / "class_counts_per_fold.csv",
    index=False
)

print("DONE")
print(f"- Total samples: {len(df)}")
print(f"- Folds: {N_SPLITS}")
print("- Full class distribution:")
print(class_counts_full)
```

# T√¨m nh·ªØng feature ·∫£nh h∆∞·ªüng nh·∫•t (feature  ‚Äúc√≥ ƒë√≥ng g√≥p‚Äù ƒë·ªÉ gi·∫£m loss)

```python



# ======================
# CONFIG
# ======================
FOLDS_PATH = Path("cv_outputs/cv_folds.pkl")
OUT_DIR = Path("xgb_importance_outputs")
OUT_DIR.mkdir(parents=True, exist_ok=True)

TARGET = "p1q9_job_current"
SEED = 42
N_CLASSES = 6
N_SPLITS = 3
TOPK_PER_FOLD = 30 # Ch·ªçn 30 feature quan tr·ªçng nh·∫•t m·ªói fold
FINAL_TOPK = 15 # Ch·ªçn 15 feature quan tr·ªçng nh·∫•t cu·ªëi c√πng


# ======================
# LOAD DATA
# ======================
df = pd.read_csv(DATA_PATH)
if TARGET not in df.columns:
    raise ValueError(f"Target '{TARGET}' not found.")

X = df.drop(columns=[TARGET])
y = df[TARGET].astype(int)
feature_names = X.columns.tolist()

# ======================
# LOAD CV FOLDS
# ======================
with open(FOLDS_PATH, "rb") as f:
    folds = pickle.load(f)

if not isinstance(folds, list) or len(folds) != N_SPLITS:
    raise ValueError("cv_folds.pkl kh√¥ng ƒë√∫ng 3 folds.")


# ======================
# TRAIN XGB PER FOLD
# ======================
fold_gain_rows = []
fold_topk = defaultdict(list)

for fd in folds:
    fold_id = int(fd["fold"])
    train_idx = fd["train_idx"]
    val_idx = fd["val_idx"]

    X_train = X.iloc[train_idx]
    y_train = y.iloc[train_idx]
    X_val = X.iloc[val_idx]
    y_val = y.iloc[val_idx]

    model = XGBClassifier(
        n_estimators=2000,          # upper bound, s·∫Ω d·ª´ng s·ªõm
        max_depth=4,                # QUAN TR·ªåNG: gi·∫£m t·ª´ 6 ‚Üí 4
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,

        # ===== REGULARIZATION (C·ª∞C K·ª≤ QUAN TR·ªåNG) =====
        min_child_weight=5,         # ch·∫∑n split theo noise
        gamma=0.5,                  # c·∫ßn gain ƒë·ªß l·ªõn m·ªõi split
        reg_alpha=0.1,              # L1 ‚Üí ·ªïn ƒë·ªãnh feature
        reg_lambda=1.0,             # L2

        objective="multi:softprob",
        num_class=N_CLASSES,
        eval_metric="mlogloss",
        random_state=SEED,
        n_jobs=-1
    )

    model.fit(
        X_train, y_train,
        eval_set=[(X_val, y_val)],
        verbose=False
    )

    booster = model.get_booster()
    gain_dict = booster.get_score(importance_type="gain")
    # gain_dict: {feature_name: gain}

    # kh·ªüi t·∫°o gain = 0 cho t·∫•t c·∫£ feature
    gains = {fname: 0.0 for fname in feature_names}

    # ghi ƒë√® gain cho feature c√≥ split
    for fname, g in gain_dict.items():
        gains[fname] = float(g)

    # l∆∞u gain t·ª´ng feature
    for fname in feature_names:
        fold_gain_rows.append({
            "fold": fold_id,
            "feature": fname,
            "gain": gains[fname]
        })

    # top K c·ªßa fold n√†y
    sorted_feats = sorted(
        gains.items(),
        key=lambda x: x[1],
        reverse=True
    )

    top_feats = sorted_feats[:TOPK_PER_FOLD]
    for rank, (fname, _) in enumerate(top_feats, start=1):
        fold_topk[fname].append(rank)


# ======================
# AGGREGATE RESULTS
# ======================
df_fold_gain = pd.DataFrame(fold_gain_rows)
df_fold_gain.to_csv(OUT_DIR / "fold_feature_gain.csv", index=False)

agg = (
    df_fold_gain
    .groupby("feature")["gain"]
    .agg(["mean", "std"])
    .reset_index()
    .rename(columns={"mean": "gain_mean", "std": "gain_std"})
)

# voting + avg rank
agg["vote_count_top15"] = agg["feature"].apply(
    lambda f: len(fold_topk.get(f, []))
)
agg["avg_rank_in_top15"] = agg["feature"].apply(
    lambda f: np.mean(fold_topk[f]) if f in fold_topk else np.nan
)

# sort: voting tr∆∞·ªõc, r·ªìi gain
agg_sorted = agg.sort_values(
    by=["vote_count_top15", "gain_mean"],
    ascending=[False, False]
).reset_index(drop=True)

agg_sorted.to_csv(
    OUT_DIR / "feature_gain_agg_3fold.csv",
    index=False
)

top15 = agg_sorted.head(FINAL_TOPK)

# ======================
# PRINT TOP 15
# ======================
print("\n===== TOP 15 FEATURES (XGB gain, 3-fold CV) =====\n")
print(
    top15[
        ["feature", "gain_mean", "gain_std", "vote_count_top15", "avg_rank_in_top15"]
    ].to_string(index=False)
)

print("\nSaved files:")
print("- fold_feature_gain.csv")
print("- feature_gain_agg_3fold.csv")
```

```python
from __future__ import annotations

from pathlib import Path
from dataclasses import dataclass
from typing import Dict, List

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.metrics import f1_score, balanced_accuracy_score, classification_report
from xgboost import XGBClassifier

import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset


# ======================
# CONFIG
# ======================
DATA_PATH = "dataset_THCS_only_filtered.csv"
OUT_DIR = Path("cvae_outputs")
OUT_DIR.mkdir(parents=True, exist_ok=True)

TARGET = "p1q9_job_current"
N_CLASSES = 6

SEED = 42
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# CVAE
VAE_EPOCHS = 120
BATCH_SIZE = 128
LR = 1e-3
BETA = 1.0
HIDDEN = 256
LATENT = 32
EARLY_STOP_PATIENCE = 12

# Only synthesize minority classes
SYN_CLASSES = [4, 5]


# ======================
# Utils
# ======================
def set_seed(seed: int = 42):
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)


# ======================
# Column grouping
# ======================
@dataclass
class ColumnGroups:
    binary: List[str]
    likert: List[str]
    categorical: List[str]
    other_numeric: List[str]


def infer_column_groups(df: pd.DataFrame, feature_cols: List[str]) -> ColumnGroups:
    binary, likert, categorical, other_numeric = [], [], [], []

    for c in feature_cols:
        s = df[c].dropna()

        if len(s) == 0:
            categorical.append(c)
            continue

        if (s < 0).any():
            categorical.append(c)
            continue

        uniq = np.sort(s.unique())

        if len(uniq) <= 2 and set(uniq).issubset({0, 1}):
            binary.append(c)
            continue

        if set(uniq).issubset({1, 2, 3, 4, 5}):
            likert.append(c)
            continue

        if pd.api.types.is_integer_dtype(s) and len(uniq) <= 20:
            categorical.append(c)
            continue

        other_numeric.append(c)

    return ColumnGroups(binary, likert, categorical, other_numeric)


# ======================
# Preprocessor
# ======================
@dataclass
class Preprocessor:
    groups: ColumnGroups
    ohe: OneHotEncoder
    scaler_likert: MinMaxScaler
    scaler_other: MinMaxScaler
    cat_feature_slices: Dict[str, slice]
    feature_names_out: List[str]

    def transform(self, df: pd.DataFrame) -> np.ndarray:
        parts = []

        if self.groups.binary:
            parts.append(df[self.groups.binary].to_numpy(np.float32))

        if self.groups.likert:
            parts.append(self.scaler_likert.transform(df[self.groups.likert]).astype(np.float32))

        if self.groups.other_numeric:
            parts.append(self.scaler_other.transform(df[self.groups.other_numeric]).astype(np.float32))

        if self.groups.categorical:
            parts.append(self.ohe.transform(df[self.groups.categorical]).astype(np.float32))

        return np.concatenate(parts, axis=1)

    def enforce_valid_onehot(self, X: np.ndarray) -> np.ndarray:
        X2 = X.copy()
        for _, sl in self.cat_feature_slices.items():
            block = X2[:, sl]
            idx = np.argmax(block, axis=1)
            hard = np.zeros_like(block)
            hard[np.arange(len(idx)), idx] = 1.0
            X2[:, sl] = hard
        return np.clip(X2, 0.0, 1.0)


def fit_preprocessor(df_train: pd.DataFrame, feature_cols: List[str]) -> Preprocessor:
    groups = infer_column_groups(df_train, feature_cols)

    scaler_likert = MinMaxScaler()
    scaler_other = MinMaxScaler()

    if groups.likert:
        scaler_likert.fit(df_train[groups.likert])
    if groups.other_numeric:
        scaler_other.fit(df_train[groups.other_numeric])

    ohe = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
    if groups.categorical:
        ohe.fit(df_train[groups.categorical])

    feature_names_out = []
    cat_feature_slices = {}
    offset = 0

    for c in groups.binary:
        feature_names_out.append(c)
        offset += 1

    for c in groups.likert:
        feature_names_out.append(f"{c}__mm")
        offset += 1

    for c in groups.other_numeric:
        feature_names_out.append(f"{c}__mm")
        offset += 1

    if groups.categorical:
        for col, cats in zip(groups.categorical, ohe.categories_):
            start = offset
            for v in cats:
                feature_names_out.append(f"{col}=={v}")
            offset += len(cats)
            cat_feature_slices[col] = slice(start, offset)

    return Preprocessor(
        groups,
        ohe,
        scaler_likert,
        scaler_other,
        cat_feature_slices,
        feature_names_out
    )


# ======================
# CVAE
# ======================
class CVAE(nn.Module):
    def __init__(self, x_dim, y_dim):
        super().__init__()
        self.enc = nn.Sequential(
            nn.Linear(x_dim + y_dim, HIDDEN),
            nn.ReLU(),
            nn.Linear(HIDDEN, HIDDEN),
            nn.ReLU()
        )
        self.mu = nn.Linear(HIDDEN, LATENT)
        self.logvar = nn.Linear(HIDDEN, LATENT)

        self.dec = nn.Sequential(
            nn.Linear(LATENT + y_dim, HIDDEN),
            nn.ReLU(),
            nn.Linear(HIDDEN, HIDDEN),
            nn.ReLU(),
            nn.Linear(HIDDEN, x_dim)
        )

    def forward(self, x, y):
        h = self.enc(torch.cat([x, y], 1))
        mu, logvar = self.mu(h), self.logvar(h)
        std = torch.exp(0.5 * logvar)
        z = mu + std * torch.randn_like(std)
        logits = self.dec(torch.cat([z, y], 1))
        return logits, mu, logvar


def train_cvae(X, y):
    X_tr, X_va, y_tr, y_va = train_test_split(X, y, test_size=0.1, stratify=y, random_state=SEED)

    def oh(a): return np.eye(N_CLASSES)[a]

    ds_tr = TensorDataset(
        torch.tensor(X_tr, dtype=torch.float32),
        torch.tensor(oh(y_tr), dtype=torch.float32)
    )
    ds_va = TensorDataset(
        torch.tensor(X_va, dtype=torch.float32),
        torch.tensor(oh(y_va), dtype=torch.float32)
    )

    dl_tr = DataLoader(ds_tr, BATCH_SIZE, shuffle=True)
    dl_va = DataLoader(ds_va, BATCH_SIZE)

    model = CVAE(X.shape[1], N_CLASSES).to(DEVICE)
    opt = torch.optim.Adam(model.parameters(), lr=LR)
    loss_fn = nn.BCEWithLogitsLoss()

    best, patience = 1e9, 0
    for _ in range(VAE_EPOCHS):
        model.train()
        for xb, yb in dl_tr:
            xb, yb = xb.to(DEVICE), yb.to(DEVICE)
            logits, mu, logvar = model(xb, yb)
            recon = loss_fn(logits, xb)
            kl = -0.5 * torch.mean(1 + logvar - mu**2 - logvar.exp())
            loss = recon + BETA * kl
            opt.zero_grad(); loss.backward(); opt.step()

        model.eval()
        with torch.no_grad():
            val_loss = 0
            for xb, yb in dl_va:
                xb, yb = xb.to(DEVICE), yb.to(DEVICE)
                logits, mu, logvar = model(xb, yb)
                recon = loss_fn(logits, xb)
                kl = -0.5 * torch.mean(1 + logvar - mu**2 - logvar.exp())
                val_loss += (recon + BETA * kl).item()

        if val_loss < best:
            best = val_loss
            best_state = model.state_dict()
            patience = 0
        else:
            patience += 1
            if patience >= EARLY_STOP_PATIENCE:
                break

    model.load_state_dict(best_state)
    model.eval()
    return model


def synthesize(model, n, cls):
    z = torch.randn(n, LATENT).to(DEVICE)
    y = torch.zeros(n, N_CLASSES).to(DEVICE)
    y[:, cls] = 1
    with torch.no_grad():
        logits = model.dec(torch.cat([z, y], 1))
    return torch.sigmoid(logits).cpu().numpy()


# ======================
# MAIN
# ======================
def main():
    set_seed(SEED)

    df = pd.read_csv(DATA_PATH)
    feature_cols = [c for c in df.columns if c != TARGET]

    df_tr, df_va = train_test_split(df, test_size=0.2, stratify=df[TARGET], random_state=SEED)
    prep = fit_preprocessor(df_tr, feature_cols)

    X_tr = prep.transform(df_tr)
    y_tr = df_tr[TARGET].values

    X_va = prep.transform(df_va)
    y_va = df_va[TARGET].values

    cvae = train_cvae(X_tr, y_tr)

    counts = pd.Series(y_tr).value_counts()
    target_n = counts.loc[[0, 1, 2, 3]].min()

    syn_X, syn_y = [], []
    for cls in SYN_CLASSES:
        need = max(0, target_n - counts.get(cls, 0))
        if need > 0:
            xs = synthesize(cvae, need, cls)
            xs = prep.enforce_valid_onehot(xs)
            syn_X.append(xs)
            syn_y.append(np.full(need, cls))

    if syn_X:
        X_syn = np.vstack(syn_X)
        y_syn = np.concatenate(syn_y)
    else:
        X_syn = np.empty((0, X_tr.shape[1]))
        y_syn = np.array([])

    # ===== SAVE FOR BAR =====
    def inverse_for_bar(X):
        out = {}
        idx = 0
        for c in prep.groups.binary:
            out[c] = X[:, idx].round().astype(int); idx += 1
        for c in prep.groups.likert:
            out[c] = np.rint(np.clip(X[:, idx], 0, 1) * 4 + 1).astype(int); idx += 1
        idx += len(prep.groups.other_numeric)
        for c, sl in prep.cat_feature_slices.items():
            cats = prep.ohe.categories_[prep.groups.categorical.index(c)]
            out[c] = np.array(cats)[np.argmax(X[:, sl], axis=1)]
        return pd.DataFrame(out)

    df_real_bar = df_tr[prep.groups.binary + prep.groups.likert + prep.groups.categorical].copy()
    df_real_bar["source"] = "real"
    df_real_bar[TARGET] = y_tr

    df_syn_bar = inverse_for_bar(X_syn)
    df_syn_bar["source"] = "synthetic"
    df_syn_bar[TARGET] = y_syn

    pd.concat([df_real_bar, df_syn_bar]).to_csv(
        OUT_DIR / "real_vs_synthetic_for_bar.csv", index=False
    )

    # ===== XGB =====
    X_aug = np.vstack([X_tr, X_syn])
    y_aug = np.concatenate([y_tr, y_syn])

    clf = XGBClassifier(
        n_estimators=800,
        max_depth=6,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        objective="multi:softprob",
        num_class=N_CLASSES,
        eval_metric="mlogloss",
        random_state=SEED
    )
    clf.fit(X_aug, y_aug)

    y_pred = clf.predict(X_va)

    print("Macro-F1:", f1_score(y_va, y_pred, average="macro"))
    print("Balanced Acc:", balanced_accuracy_score(y_va, y_pred))

    pd.DataFrame(classification_report(y_va, y_pred, output_dict=True)).T.to_csv(
        OUT_DIR / "classification_report.csv"
    )


if __name__ == "__main__":
    main()
```

# üìâ ƒê√°nh Gi√° Th·ª±c Nghi·ªám: H·∫°n Ch·∫ø C·ªßa CVAE Oversampling

### 1. Hi·ªán tr·∫°ng
M·∫∑c d√π ƒë√£ √°p d·ª•ng k·ªπ thu·∫≠t sinh d·ªØ li·ªáu (Data Generation) b·∫±ng CVAE, k·∫øt qu·∫£ hu·∫•n luy·ªán cho th·∫•y m√¥ h√¨nh **v·∫´n kh√¥ng th·ªÉ nh·∫≠n di·ªán ƒë∆∞·ª£c L·ªõp 4 v√† L·ªõp 5**.

### 2. Ph√¢n t√≠ch nguy√™n nh√¢n (Root Cause Analysis)
Ph∆∞∆°ng ph√°p CVAE oversampling ƒë∆∞·ª£c ƒë√°nh gi√° l√† **"ch∆∞a t·ªët"** trong tr∆∞·ªùng h·ª£p c·ª• th·ªÉ n√†y do s·ª± khan hi·∫øm d·ªØ li·ªáu c·ª±c ƒëoan:

* **ƒê·∫ßu v√†o kh√¥ng ƒë·ªß (Insufficient Input):** L·ªõp 5 ch·ªâ t·ªìn t·∫°i **02 m·∫´u** trong t·∫≠p hu·∫•n luy·ªán (training set). ƒê√¢y l√† con s·ªë qu√° nh·ªè ƒë·ªÉ b·∫•t k·ª≥ m√¥ h√¨nh Generative n√†o c√≥ th·ªÉ h·ªçc ƒë∆∞·ª£c ph√¢n ph·ªëi x√°c su·∫•t.
* **Ch·∫•t l∆∞·ª£ng d·ªØ li·ªáu sinh (Generated Data Quality):**
    * Do kh√¥ng gian m·∫´u qu√° h·∫πp, m√¥ h√¨nh b·ªã r∆°i v√†o tr·∫°ng th√°i *Overfitting* ho·∫∑c *Mode Collapse*.
